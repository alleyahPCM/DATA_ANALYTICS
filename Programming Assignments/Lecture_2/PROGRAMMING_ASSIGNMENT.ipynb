```python
dropout_students = data[data['Target'] == 'Dropout']
total_students = len(data)

# Calculate dropout rate
dropout_rate = (len(dropout_students) / total_students) * 100
print("Dropout Rate: {:.2f}%".format(dropout_rate))
```

```python
dropout_by_course = dropout_students.groupby('Course').size().reset_index(name='Dropouts')

dropout_by_course = dropout_by_course.sort_values(by='Dropouts', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(dropout_by_course['Course'], dropout_by_course['Dropouts'], color='skyblue')
plt.title('Number of Dropouts per Course')
plt.xlabel('Course')
plt.ylabel('Number of Dropouts')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
